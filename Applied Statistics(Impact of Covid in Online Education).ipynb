```python
def Plot_NumberOfResponses(df):
  """
  This function used to get the number of responses on each day, hour and in day of week.
  Input : df - dataframe 
  Plots : Number of responses each day 
  """
  dataset=df.copy()
  dataset['day']=pd.DatetimeIndex(dataset['Timestamp']).day
  dataset['day_of_week'] = pd.DatetimeIndex(dataset['Timestamp']).day_name()
  dataset['hour']=pd.DatetimeIndex(dataset['Timestamp']).hour
  fig, ((ax1,ax2,ax3)) = plt.subplots(nrows=1,ncols=3,figsize=(25,7)) # figure of 2 rows and 3 columns subplots 
  # count plots 
  s1=sns.countplot(x="day",data=dataset,ax=ax1,palette='dark:salmon_r')
  s2=sns.countplot(x="hour",data=dataset,ax=ax2,palette='magma')
  s3=sns.countplot(x="day_of_week",data=dataset,ax=ax3,palette='crest')

  # annotating the values in plots 
  for p in s1.patches:
    s1.annotate(format(p.get_height(), ''), (p.get_x() + p.get_width()/2 , p.get_height()), ha = 'center', va = 'center', xytext = (0,10),rotation=0, textcoords = 'offset points',fontsize=12)
  for p in s2.patches:
    s2.annotate(format(p.get_height(), ''), (p.get_x() + p.get_width()/2 , p.get_height()), ha = 'center', va = 'center', xytext = (0,10),rotation=0, textcoords = 'offset points',fontsize=12)
  for p in s3.patches:
    s3.annotate(format(p.get_height(), ''), (p.get_x() + p.get_width()/2 , p.get_height()), ha = 'center', va = 'center', xytext = (0,10),rotation=0, textcoords = 'offset points',fontsize=12)
 
  ax1.title.set_text("Number of Responses Vs day")
  ax2.title.set_text("Number of Responses Vs hour")
  ax3.title.set_text("Number of Responses Vs day_of_week")

  plt.show()
```

##### **Modifying College/University feature -> College - IIT H, Other IIT or Other Universities :**

```python
def College_Data_Modifier(df,col):
  """
  This function changes the college names to either IIT H , Other IIT or other Universities depends on the college given 
  Input : df - given dataframe 
          col - column name 
  Returns : dataset with given functionality added 
  """
  # copying the dataframe 
  dataset=df.copy()
  # getting the list of college names 
  college_data=list(dataset[col])
  # initializing the output list 
  expected_result=[]
  # checking the conditions and updating the list 
  for c_name in college_data:
    # remove ","
    colg=""
    for ele in c_name:
      if ele!=",":
        colg+=ele 
    # getting lower case 
    colg=colg.lower()
    # split based on space 
    colg=colg.split(" ")
    # conditional check 
    if colg[0]=='iith' or colg[0]=='iit-h' or colg[0]=="iit-hyderabad" or colg[0]=="iithyd":
      expected_result.append("IIT H")
    elif colg[0]=='iit':
      if len(colg)>1:
        if colg[1][0]=='h':
          expected_result.append("IIT H")
        else:
          expected_result.append("Other IIT")
      else:
        expected_result.append("Other IIT")
    elif colg[0]=="iit":
      if "hyderabad" in colg:
        expected_result.append("IIT H")
      else:
        expected_result.append("Other IIT")
    elif colg[0]=='indian':
      if 'hyderabad' in colg:
        expected_result.append("IIT H")
      elif "information" not in colg:
        expected_result.append("Other IIT")
      else:
        expected_result.append("Other colleges")
    else:
      expected_result.append("Other colleges")
  # create a new column and add the updated list to the dataframe 
  dataset[col]=expected_result
  return dataset
```

```python
import statistics as tools
from scipy import stats
import math
```

```python
def estimate_SingleMeanCI(df, alpha):
  """
  Function used to estimate the Confidence Interval of single mean based on the given alpha value (Theorem 2)
  population variance unknown
  Input: df - dataframe
         alpha - level of significance (2 tails)
  Note : The distribution must be normal and has to be verified before sending to this function. 
  """
  sample_Size = df.shape[0]
  sample_Mean = tools.mean(df)
  sample_Variance = tools.stdev(df)
  # Since t.ppf gives the lower tail probabiity, for alpha significance, 1-(alpha/2) is considered
  t_value = stats.t.ppf(1-(alpha/2), sample_Size-1)
  LL = (sample_Mean)-((t_value)*(sample_Variance/math.sqrt(sample_Size)))
  UL = (sample_Mean)+((t_value)*(sample_Variance/math.sqrt(sample_Size)))
  return (LL,UL)
```

```python
hours_Df =  Preprocessed_Survey_Data['Total Hours Spent(class, assignment and self-study)']
```

```python
stats.t.ppf(1-0.025, 7)
```

```python
# Alphas for confidence intervals corresponding to range [90,99] percentage respectively 
CI_Alphas_list = np.linspace(0.1, 0.01, 10)
```

```python
lower_bound = []
upper_bound = []
confidence_Length_list = []
for i in CI_Alphas_list:
  CI_percent = (1-i)*100
  confidence_Interval = estimate_SingleMeanCI(hours_Df, i)
  print(CI_percent,"% confidence interval on mean number of hours per day = ", confidence_Interval)
  lower_bound.append(confidence_Interval[0])
  upper_bound.append(confidence_Interval[1])
  confidence_Length = confidence_Interval[1]-confidence_Interval[0]
  print("Confidence length = ", confidence_Length)
  confidence_Length_list.append(confidence_Length)
  print("-----------------------------------------------")
```

```python
conf_df = pd.DataFrame()
conf_value = []
for i in range(len(CI_Alphas_list)):
  CI_percent = (1-CI_Alphas_list[i])*100
  conf_value.append("Confidence_"+str(int(CI_percent)))
conf_df['Confidence Level']=conf_value
conf_df['Lower Bound']=lower_bound
conf_df['Upper Bound']=upper_bound
```

```python
plt.figure(figsize=(10,6))

for lower,upper,y in zip(conf_df['Lower Bound'],conf_df['Upper Bound'],range(len(conf_df))):
  plt.plot((lower,upper),(y,y),'ro-',color='red')
plt.yticks(range(len(conf_df)),list(conf_df['Confidence Level']))

# true mean line 
plt.axvline(tools.mean(hours_Df),label="Sample mean")
plt.xlabel("Confidence Level")
plt.ylabel("Hours")
plt.title("CI estimations for mean amount of studying hours per day")
plt.legend()
```

Now both the distributions are normal and confidence interval for difference between means can be calculated

```python
def get_WelchIntervalMargin(df1,df2,alpha):
  """
  Function to calculate the Welch's t-interval margin
  Input: df1 - dataframe-1
         df2 - dataframe-2
  """
  df1_sampleVar = tools.variance(df1)
  df2_sampleVar = tools.variance(df2)
  df1_sampleSize = df1.shape[0]
  df2_sampleSize = df2.shape[0]
  temp = df1_sampleVar/df1_sampleSize + df2_sampleVar/df2_sampleSize
  numerator = math.pow(temp,2)
  denominator1 = (math.pow((df1_sampleVar/df1_sampleSize),2))/df1_sampleSize-1
  denominator2 = (math.pow((df2_sampleVar/df2_sampleSize),2))/df2_sampleSize-1
  r = numerator/(denominator1 + denominator2)
  # Since t.ppf gives the lower tail probabiity, for alpha significance, 1-(alpha/2) is considered
  t_value = stats.t.ppf(1-(alpha/2), r)
  intervalMargin = t_value*math.sqrt(temp)
  return intervalMargin

def get_IntervalMargin(df1, df2,alpha):
  """
  Function to calculate the confidence interval margin for equal variances
  Input: df1 - dataframe-1
         df2 - dataframe-2
  """
  df1_sampleVar = tools.variance(df1)
  df2_sampleVar = tools.variance(df2)
  df1_sampleSize = df1.shape[0]
  df2_sampleSize = df2.shape[0]
  deg_freedom = df1_sampleSize+df2_sampleSize-2
  pooled_var = (df1_sampleVar*(df1_sampleSize-1)+df2_sampleVar*(df2_sampleSize-1))/deg_freedom
  # Since t.ppf gives the lower tail probabiity, for alpha significance, 1-(alpha/2) is considered
  t_value = stats.t.ppf(1-(alpha/2), deg_freedom)
  temp = 1/df1_sampleSize + 1/df2_sampleSize
  intervalMargin = t_value*math.sqrt(pooled_var)*math.sqrt(temp)
  return intervalMargin
```

```python
def estimate_DifferenceMeanCI(df1, df2, alpha):
  """
  Function used to estimate the Confidence Interval of difference means based on the given alpha value (Theorem 3,4)
  population variance unknown
  Input: df1 - dataframe-1
         df2 - dataframe-2  
         alpha - level of significance (2 tails)
  Note : The distributions must be normal and has to be verified before sending to this function. 
  """
  df1_sampleMean = tools.mean(df1)
  df2_sampleMean = tools.mean(df2)
  df1_sampleVar = tools.variance(df1)
  df2_sampleVar = tools.variance(df2)
  df1_sampleSize = df1.shape[0]
  df2_sampleSize = df2.shape[0]
  if (df1_sampleVar/df2_sampleVar>4) or (df2_sampleVar/df1_sampleVar>4):
    margin = get_WelchIntervalMargin(df1,df2,alpha)
  else:
    margin = get_IntervalMargin(df1, df2,alpha)
  LL = df1_sampleMean-df2_sampleMean-margin
  UL = df1_sampleMean-df2_sampleMean+margin
  return (LL,UL)
```
